```python
import cv2
import numpy as np
```

```python
IDphoto = cv2.imread(r"C:\Users\imyla\Downloads\Profile.jpg")
BarCode = cv2.imread(r"C:\Users\imyla\Downloads\barcodes.png")
```

```python
card = np.full((250,450,3),255,dtype = np.uint8)
```

```python
IDphoto = cv2.resize(IDphoto,(160,160))
BarCode = cv2.resize(BarCode, (300,50))
```

```python
print(card.shape)
print(BarCode.shape)
```

```python
card[20:180 , 20:180]= IDphoto
card[190:240, 60:360] = BarCode
```

```python
card = cv2.rectangle(card,(400,0),(450,250),(139, 0, 0),-1)
```

```python
name = input("Name: ")
ID = input("ID: ")
Dep = input("Department: ")
Pos = input("Position: ")
```

```python
card = cv2.putText(card,"ABC Company",(190,40),cv2.FONT_HERSHEY_COMPLEX,1,(139, 0, 0),2)
card = cv2.putText(card,"Name: "+ name,(220,90),cv2.FONT_HERSHEY_COMPLEX,0.5,(0,0,0),1)
card = cv2.putText(card,"ID: "+ ID,(220,110),cv2.FONT_HERSHEY_COMPLEX,0.5,(0,0,0),1)
card = cv2.putText(card,"Department: "+ Dep,(220,130),cv2.FONT_HERSHEY_COMPLEX,0.5,(0,0,0),1)
card = cv2.putText(card,"Position: "+ Pos,(220,150),cv2.FONT_HERSHEY_COMPLEX,0.5,(0,0,0),1)
```

```python
cv2.imshow("ID card",card)
cv2.waitKey(0)
```
